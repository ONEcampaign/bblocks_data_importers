# WEO Importer

The `WEO` importer is used to import data from the World Economic Outlook (WEO) database.

### Basic Usage

```python

from bblocks import data_importers as bbdata
```

```python
weo = bbdata.WEO() # initialize the WEO importer
df = weo.get_data() # get the data as a pandas DataFrame

df.head()
```

By default the `get data` method will retrieve the latest available published version of the World Economic Outlook. You can also specify a specific version. Note: the World Economic Outlook is published twice a year in April and October.

```python
df = weo.get_data(version=("April", 2024))
df.head()
```

The imported data is cached in the object to avoid repeated downloads. At any time you can clear the cache by calling the `clear_cache` method.

```python
weo.clear_cache()
```
